```python
import pandas as pd 
import numpy as np 
import random as rnd
import matplotlib.pyplot as plt 
import nltk
```

```python
tr_data = pd.read_csv('Data/train_data.csv')
```

```python
tr_true = tr_data['text'][tr_data['target'] == 1]
tr_false = tr_data['text'][tr_data['target'] == 0]
tr_true = tr_true.map(lambda sentence: nltk.pos_tag( nltk.word_tokenize(sentence)))
```

```python
tr_false = tr_false.map(lambda sentence: nltk.pos_tag( nltk.word_tokenize(sentence)))
```

```python
tst_data = pd.read_csv('Data/test_data.csv')
tst_true = tst_data['text'][tst_data['target'] == 1]
tst_false = tst_data['text'][tst_data['target'] == 0]
test_tts = {'real_trag' : [w for w in tst_true.to_list() if len(w) > 1 and w not in {'in', 'out', 'at', 'under', 'over','on', 'after', 'bellow', 'aside', 'behind', 'close', }], 'not_trag' : tst_false.to_list() } 

#retirar as palavras que aparecem menos do que duas vezes. 
```

```python
to_keep = {'VB', 'VBG', 'VBN', 'VBZ' , 'VBP'}
clean_tr_true = [] 
for sentence in tr_true:
    selected = []
    for w in sentence:
        if w[1] in to_keep:
            selected.append(w[0].lower())
    if selected != []:
        clean_tr_true.append(selected)
```

```python
vocab_words_true_dis = [] 

for sentence in clean_tr_true:
    for word in sentence:
        vocab_words_true_dis.append(word)
```

```python
vocab_unique_words_true_dis = list(dict.fromkeys(vocab_words_true_dis))
```

```python
vocab_unique_words_true_dis[:5]
```

##### Fazendo um dicionario de onde cada palavra única que conhecemenos nos nossos dados de treinamento vai ter como chave um coeficiente que representa a chance dela aparecer em um twitter sobre desastre real
Esse coeficiente nada mais é do que a probabilidade de determinada palavra ocorrer, considerando que as palavras são independentes (sabemos que isso não é verdade) e considerando que os tweets independentes

```python
max_prob_w = ''
max_prob = -1
```

```python
dict_prob_true = {}
for w in vocab_unique_words_true_dis:
    if len(w) > 1:
        tts_with_w = 0 
        for sentence in clean_tr_true:
            if w in sentence:
                tts_with_w += 1 
        print(f"Number of real disaster tweets with the word {w}: {tts_with_w}")
        total_true_dis = len(clean_tr_true)
        prob = (tts_with_w + 1)/(total_true_dis + 2)
        print(f"Probability of {w} is {prob} \n") 
        dict_prob_true[w.lower()] = prob * 100
        if prob > max_prob:
            max_prob = prob 
            max_prob_w = w.lower() 
```

```python
print(f'The most prob word is {max_prob_w} : {max_prob}')
```

```python

true_sample = list(dict_prob.items())
rnd.shuffle(true_sample) 
true_sample = dict(true_sample[:10])
```

```python
plt.figure(figsize=(10, 6), dpi=80)
plt.bar(*zip(*true_sample.items()))
plt.show()
```

```python
dict_prob_false = dict( {} ) 


to_keep = {'VB', 'VBG', 'VBN', 'VBZ' , 'VBP'}
clean_tr_false = [] 
for sentence in tr_false:
    for w in sentence:
        if w[1] in to_keep and w[0] != '' and len(w[0])>1:
            clean_tr_false.append(w[0].lower())
```

```python
len(clean_tr_false)
```

```python
clean_tr_false_unique = list(set(clean_tr_false))
len(clean_tr_false_unique)
```

```python
dict_word_false = {}
```

```python
max_w_false = ''
max_prob_false = -1
for w in clean_tr_false_unique:
    count = 0 
    for word in clean_tr_false:
        if w == word:
            count += 1 
    prob = count/len(clean_tr_false)
    prob = prob*100
    if prob > max_prob_false:
        max_w_false = w 
        max_prob_false = prob 
    print(f'{w} prob: {prob}')
    dict_word_false[w] = prob 

    
```

```python
print(f'{max_w_false} : {max_prob_false}')
```

```python

false_sample = list(dict_prob_false.items())
rnd.shuffle(false_sample) 
false_sample = dict(false_sample[:10])
```

```python
plt.bar(*zip(*false_sample.items()))
plt.show()
```
